```python
import pandas as pd

from ast import literal_eval
import ast
```

```python
df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
#1.Drop columns
df_sub = df.drop(columns=["homepage", "id", "original_language", "original_title","overview", "production_countries", \
                 "spoken_languages", "status", "tagline", "title"])
```

```python
#2. drop NULL
df_sub = df_sub.dropna()
df_sub.shape
```

```python
sum(df_sub.budget == 0)
```

```python
df_sub = df_sub.loc[~(df['budget'] == 0) ]
```

```python
df_sub.shape
```

```python
df.genres[0]
```

```python
#convert dict to list
df_sub['genres'] = df_sub['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
df_sub['keywords'] = df_sub['keywords'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
df_sub['production_companies'] = df_sub['production_companies'].apply(ast.literal_eval)
df_sub['production_companies'] = df_sub['production_companies'].fillna("[]").apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df_sub.head()
```

```python
#create dummy vatiables for genre
df_genre = df_sub.genres.str.join('|').str.get_dummies().add_prefix('genres_')
df_sub = pd.concat([df_sub, df_genre], axis=1)
df_sub.head(2)
```

```python
df_genre.head()
```

```python
from collections import defaultdict
production = defaultdict(int)

for col in df_sub["production_companies"]:
    for i in col:
        production[i] +=  1

sorted_production = sorted(production.items(), key=lambda kv: kv[1], reverse=True)
len(sorted_production)
```

```python
sorted_production
```

```python
#create dummy variables for productions
df_production = df_sub.production_companies.str.join('|').str.get_dummies().add_prefix('productions_')
#df_production = df_production.drop(columns=[col for col in df_production if col not in top_10_production])
```

```python
df_production.shape
```

```python
df_sub = pd.concat([df_sub, df_production], axis=1)
df_sub.head(2)
```

```python
df_sub.shape
```

```python
from collections import defaultdict
genre = defaultdict(int)

for col in df_sub["genres"]:
    for i in col:
        genre[i] +=  1

sorted_genre = sorted(genre.items(), key=lambda kv: kv[1])
sorted_genre        
```
